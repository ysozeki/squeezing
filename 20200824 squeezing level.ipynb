```python
import numpy as np
import scipy.special as sp 
import matplotlib.pyplot as plt
%matplotlib inline

plt.matplotlib.rc('text', usetex = True)
plt.matplotlib.rc('grid', linestyle = 'dotted')
plt.matplotlib.rc('figure', figsize = (6.4,4.8)) # (width,height) inches

x = np.linspace(0, 2, 500)

plt.plot(x, 20*np.log10(np.exp(np.sqrt(x))))
plt.plot(x, 20*np.log10(np.cosh(np.sqrt(x))))
plt.plot(x, 20*np.log10(np.sinh(np.sqrt(x))))

plt.xlim((0, 2))
plt.ylim((-5, 10))
plt.legend(('$20\log(\exp\sqrt x)$','$20\log(\cosh\sqrt x)$', '$20\log(\sinh\sqrt x)$'), loc = 0)
plt.xlabel('Normalized power')
plt.ylabel('Gain [dB]')
plt.grid(True)
plt.tight_layout(0.5)
plt.savefig('20200824 squeezing level.pdf')
```
